```python
import pandas as pd
data  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/startupdata/training_set_label.csv" )
```

```python
data.head(n=10)
```

```python
status = data['status']
```

```python
status_d = pd.get_dummies(status,drop_first=1)
```

```python
status_d.shape
```

```python
import numpy as np
```

```python

final=pd.get_dummies(data['status'])
data=pd.concat([data,final],axis=1)
data.drop('status',axis=1,inplace=True)
```

```python

data.head(n=10)
```

```python
data
```

```python
m=data['age_first_milestone_year'].mean()
data['age_first_milestone_year']=data['age_first_milestone_year'].fillna(m)
data['age_first_milestone_year'].head(n=50)
md=data['age_last_milestone_year'].mean()
data['age_last_milestone_year']=data['age_last_milestone_year'].fillna(md)
```

```python
#select numerical columns required for prediction


x=data[['latitude','longitude','avg_participants','relationships','is_top500','age_first_funding_year','age_last_funding_year','age_first_milestone_year','age_last_milestone_year','has_VC','has_angel','has_roundA','has_roundB','has_roundC','has_roundD']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split( x, status_d, test_size = 0.2, random_state = 42 )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_p = model.predict(x_test)
```

```python
y_p
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_p)
```
